```python
# Averaging the values of the columns
average_array = []

for column in df.columns:
    average_array.append(df[column].mean())

average_array
```

```python
comparison_matrix = np.zeros((16, 16))
comparison_matrix
```

```python
# Fill the matrix based on rank differences
n = 16
for i in range(n):
    for j in range(n):
        if i != j:
            #rank_difference = average_array[j] - average_array[i]
            #rank_difference = abs(rank_difference)
            #comparison_matrix[i, j] = 1 / rank_difference if rank_difference != 0 else 0
            comparison_matrix[i, j] = average_array[i] / average_array[j]
        if i == j:
            comparison_matrix[i, j] = 1

comparison_matrix = pd.DataFrame(comparison_matrix)
comparison_matrix
```

```python
# Normalizing the matrix
column_sums = np.sum(comparison_matrix, axis=0)
normalized_matrix = comparison_matrix / column_sums
normalized_matrix
```

```python
# Getting the priority vector
priority_vector = np.mean(normalized_matrix, axis=1)
priority_vector
```

```python
def calculate_consistency(matrix, priority_vector):
    n = matrix.shape[0]
    weighted_sum = np.dot(matrix, priority_vector)
    lambda_max = np.sum(weighted_sum / priority_vector) / n
    CI = (lambda_max - n) / (n - 1)
    # Assuming RI for n=16 is approximately 1.61
    RI = 1.61
    CR = CI / RI
    return CI, CR, lambda_max
```

```python
# Initialize a matrix of zeros for win counts
n_statements = len(df.columns)
win_matrix = np.zeros((n_statements, n_statements))

# Iterate through each rater's rankings and update win counts
for index, row in df.iterrows():
    rankings = row.values
    for i in range(n_statements):
        for j in range(n_statements):
            if rankings[i] < rankings[j]:  # Statement i wins against statement j
                win_matrix[i, j] += 1
np.fill_diagonal(win_matrix, 16)

win_df = pd.DataFrame(win_matrix, columns=df.columns, index=df.columns)

win_df  
```
